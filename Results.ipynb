```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

deepq = []
regq = []
random = []

with open('output/deepq-envCartPole-v1-eps500-tests5-testinterval10-time1642599926') as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        deepq.extend(row)
        
with open('output/random-envCartPole-v1-eps500-tests5-testinterval10-time1642605032') as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        random.extend(row)
        
with open('output/regq-envCartPole-v1-eps500-tests5-testinterval10-time1642604591') as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        regq.extend(row)
        
deepq = [float(x) for x in deepq]
regq = [float(x) for x in regq]
random = [float(x) for x in random]
```

```python
plt.plot(deepq)
plt.plot(regq)
plt.plot(random)
plt.ylabel("reward")
plt.xlabel("nr epochs %10")
plt.legend(["deep Q-learning", "Q-learning", "Random"])
plt.title("")
plt.show()
```
